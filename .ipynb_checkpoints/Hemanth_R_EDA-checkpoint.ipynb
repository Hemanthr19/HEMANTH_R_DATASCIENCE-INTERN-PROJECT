```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#load ds
customer = pd.read_csv("Customers.csv")
product = pd.read_csv("Products.csv")
transaction = pd.read_csv("Transactions.csv")
#display rows
print(customer.head(10))
print(product.head(10))
print(transaction.head(10))
print(customer.tail(10))
print(product.tail(10))
print(transaction.tail(10))
```

```python
#data clng and pr-prcng
#miss value
print(customer.isnull().sum())
print(product.isnull().sum())
print(transaction.isnull().sum())
print(customer.duplicated().sum())
print(product.duplicated().sum())
print(transaction.duplicated().sum())
customer["SignupDate"]=pd.to_datetime(customer["SignupDate"])
transaction["TransactionDate"]=pd.to_datetime(transaction["TransactionDate"])
```

```python
#eda
print(transaction.describe())
print(customer["Region"].value_counts())
print(product["Category"].value_counts())
print("different customer:",transaction["CustomerID"].nunique())
print("variety products:",transaction["ProductID"].nunique())
print("total transaction:",transaction.shape[0])
```

```python
#viz and insight-1
#Sales Trend Over Time
transaction.groupby("TransactionDate")["TotalValue"].sum().plot(figsize=(25,9),marker="o", linestyle="-",color="green")
plt.title("Sales over time")
plt.xlabel("Date")
plt.ylabel("Total sales")
plt.show()
```

```python
#insight 1
monthly_sales = transaction.resample("ME", on="TransactionDate")["TotalValue"].sum()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales, marker="o", linestyle="-", color="green")

plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#Top Selling Products
#insight 2
top_prod = transaction.groupby("ProductID")["Quantity"].sum().nlargest(10)
top_prod.plot(kind="bar", figsize=(12, 6))
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product ID")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
#Revenue by Category
#insight 3
merged_data = transaction.merge(product, on="ProductID")
revenue_by_category = merged_data.groupby("Category")["TotalValue"].sum()
revenue_by_category.plot(kind="bar", figsize=(12, 6),color="blue")
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#insight 4
#Customer Signups by Region
customer["Region"].value_counts().plot(kind="pie", figsize=(7, 7), autopct="%1.1f%%", colors=["blue", "green", "red", "purple"])

plt.title("Customer Distribution by Region")
plt.show()
```

```python
#avg spending per customer
avg_spending = transaction.groupby("CustomerID")["TotalValue"].sum().mean()
print(f"Average Customer Spending: ₹{avg_spending:.2f}")
```
